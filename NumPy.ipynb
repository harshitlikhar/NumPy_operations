```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Highest Holywood Grossing Movies.csv")
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
print(len(df['Movie Runtime'][0]))
h = int(df['Movie Runtime'][0][:1])
m = int(df['Movie Runtime'][0][5:7])

print(h,m)

print((h*60)+m) 
```

```python
def min_(x):
    if len(x) == 10:
        hr = int(x[:1])
        min__ = int(x[5:6])
    else:
        hr = int(x[:1])
        min__ = int(x[5:7])
    return (hr * 60) + int(min__)
```

```python
def hour(x):
    for i in range(len(x)):
        if x[i] == 'h':
            return int(x[:i])
hour(df['Movie Runtime'][0])
```

```python
df['hour'] = df['Movie Runtime'].apply(hour)
df.head()
```

```python
def min_min(x):
    for i in range(len(x)-4,0,-1):
        l = len(x)-4
        if x[i] == 'r':
            return int(x[i+1:l])
min_min(df['Movie Runtime'][0])
```

```python
df['Minutes'] = df['Movie Runtime'].apply(min_min)
df.head()
```
